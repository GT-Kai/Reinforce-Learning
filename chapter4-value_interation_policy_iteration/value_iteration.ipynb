# Value Iteration

Value iteration is used to solve the Bellman optimality equation.

the Bellman optimality equation by matrix-vector form:

$$
V = \max_{\pi} \left( \mathbf{r} + \gamma P V \right)
$$


The iteration algorithm by matrix-vector form is :

$$
V_{k+1} = \max_{\pi} \left( r_\pi + \gamma P_\pi V_k \right),  k = 1,2,3,......
$$

It can be decomposed(分解) to two steps:

Step1: Policy Update

$$
\pi_{k+1} = \argmax_{\pi} ( r_\pi + \gamma P_\pi V_k )
$$

Step2: Value Update

$$
V_{k+1} = r_{\pi_{k+1}} + \gamma P_{\pi_{k+1}} V_k
$$

Note that, $V_{k}$ is not a state value, because it's not ensured the Bellman equation.

The Bellman optimality equation by elementwise form is :

$$
v_\pi(s)
= \sum_{a} \pi(a|s) \left[ \sum_{r} P(r|s,a) r + \gamma \sum_{s'} P(s'|s,a) v_\pi(s') \right]
$$

It can be decomposed to two steps:

Step1: Policy Update

$$
\pi_{k+1} = \argmax_\pi \left( \sum_a \pi(a|s) \left[ \sum_{r} P(r|s,a) r + \gamma \sum_{s'} P(s'|s,a) v_{k} (s') \right] \right)

Step2: Value Update

$$
v_{k+1} (s) = \sum_{a} \pi_{k+1}(a|s) \left[ \sum_r P(a|s,a) r + \gamma \sum_{s'} P(s'|s,a) v_{k} (s') \right]
$$
